```python
from sklearn.metrics import average_precision_score
from time import time
import numpy as np

def map_metric(pred, true):
    n = np.min([len(pred),len(true)])
#     print(len(pred))
    matches = [1 if pred[i] == true[i] else 0 for i in range(n)]
#     print(matches)
#     print(pred)

    total_score = 0
    total_correct = 0
    for i in range(len(matches)):
        if matches[i] == 1:
            total_correct += 1
            total_score += (total_correct/(i+1))
    N = len(true)
    Y = N- total_score
    kendall = (total_correct - Y) / (total_correct + Y)
    try:
        return total_score/total_correct, kendall
    except:
        return 0, kendall
    
# map_metric([1,2,3],[3,2,1])
kendalls = []
def average_precision(true_list, predicted_list, k=40):
    true_set = frozenset(true_list)
    predicted_list = predicted_list[:k]
    precisions = []
    for i,doc_id in enumerate(predicted_list):        
        if doc_id in true_set:
            prec = (len(precisions)+1) / (i+1)            
            precisions.append(prec)
    if len(precisions) == 0:
        return 0.0
    Y = 40 - len(precisions)
    
    kendalls.append((len(precisions)-Y)/(len(precisions)+Y))
    return round(sum(precisions)/len(precisions),3)

import requests
# from time import time
# url = 'http://35.232.59.3:8080'
# place the domain you got from ngrok or GCP IP below. 
url = 'http://f76a-104-198-177-27.ngrok.io'

qs_res = []
for q, true_wids in queries.items():
  duration, ap = None, None
  t_start = time()
  try:
    res = requests.get(url + '/search', {'query': q}, timeout=35)
    duration = time() - t_start
    if res.status_code == 200:
      pred_wids, _ = zip(*res.json())
      ap = average_precision(true_wids, pred_wids)
  except:
    pass
  
  qs_res.append((q, duration, ap))

def evaluate(search):
    times = []
    maps = []
    kendalls = []
    c = 0
    for q, true_res in queries.items():
        print(c+1)
        c+=1

        t = time()
        pred_search = search(q)
        if pred_search == []:
            print("empty:",c-1)
        else:
    #         print(pred_search)
            times.append(time()-t)
            MAP, kend = average_precision([x[0] for x in pred_search], true_res)
            maps.append(MAP)
            kendalls.append(kend)
    return times,maps,kendalls
# MAP = np.mean(maps)
# maps
# times,maps,kendalls = evaluate(engine.search)
```

```python
all_maps = {}
def search_func(t,b,a):
    def sf(q):
        query_tokens = engine.index.tokenize(q)
        title_scores = engine.title_score(query_tokens)  # convert to format: {query_id: list of (doc_id,score)}
        body_scores = engine.search_body_bm25(query_tokens)  # convert to format: {query_id: list of (doc_id,score)}
        anchor_scores = dict(engine.anchor_score(query_tokens))
        return engine.merge_results(title_scores, body_scores,anchor_scores, title_weight=t, text_weight=b,anchor_weight=a, pageRank_weight=0.01, N=100)
    return sf

for anchor_w in np.arange(0.1,1.5,0.1):
    for body_w in np.arange(0.1,1.5,0.1):
        for title_w in np.arange(1.5,0.4,-0.1):
            curr_search_func = search_func(title_w,body_w,anchor_w)
            times,maps,kendalls = evaluate(curr_search_func)
            all_maps[(title_w,body_w,anchor_w)] = np.mean(maps)
```
